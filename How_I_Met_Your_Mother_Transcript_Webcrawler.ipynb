```python
import requests
from lxml import etree
import re
from bs4 import BeautifulSoup
import xlwt
from urllib import robotparser,parse 
import numpy as np
import csv
import time
import pandas as pd
```

```python
def get_html(url):
    #模擬訪問頁面的函數
    try:
        user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.131 Safari/537.36'
        resp = requests.get(url, headers={'User-Agent': user_agent}, timeout = 30,verify = False) #回傳為一個request.Response的物件
        resp.endcoding = 'utf8'
        #response = resp.content.decode()
        c = resp.status_code
        return resp.text
    except:
        return "ERROR"
```

```python
def get_url_list():
    url_list = []
    page_list = []
    for i in range(9):
        http = "http://transcripts.foreverdreaming.org/viewforum.php?f=177&sid=baf380e06a24a4fc21c538ccc65eb235&start="+str(i*25)
        page_list.append(http)
    for i in page_list:
        text_per_page = get_html(i)
        tree_per_page = etree.HTML(text_per_page)
        for j in range(6, 31):
            url = tree_per_page.xpath('//*[@id="pagecontent"]/div[1]/div[2]/table/tr[' + str(j+1) + ']/td[1]/h3/a//@href')
            if url != []:
                url_list.append("http://transcripts.foreverdreaming.org"+ url[0].strip("."))
    return url_list
```

```python
k = get_url_list()
print(k)
```

```python
def get_content(url):
    text = get_html(url)
    tree = etree.HTML(text)
    title = tree.xpath('//*[@id="pagecontent"]/div[1]/div[1]/h2/text()')
    transcript = tree.xpath('//*[@class="postbody"]//text()')
    
    transcript_neat = []
    for i in transcript:
        if i[0] != "\n":
            transcript_neat.append(i)
    
    return title, transcript_neat
    
```
